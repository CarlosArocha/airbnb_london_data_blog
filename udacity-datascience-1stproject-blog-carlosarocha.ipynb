Let's start with our first question: **WHAT** kind of properties is the most popular in airbnb London?.

```python
# We first create a dataframe with the 'room_type' of every year
df_Room17 = df_l17.room_type.value_counts().to_frame('RoomType_2017')
df_Room18 = df_l18.room_type.value_counts().to_frame('RoomType_2018')
df_Room19 = df_l19.room_type.value_counts().to_frame('RoomType_2019')
df_Room20 = df_l20.room_type.value_counts().to_frame('RoomType_2020')

df_Rooms = pd.concat([df_Room17,df_Room18,df_Room19,
                      df_Room20],axis = 1).fillna(0)

# Then we create two new columns:
#    var = Variation (%) during the 2017-2020 period of each kind of room type.
#    var_lastyear = Variation (%) during the last year (COVID year)

df_Rooms['var'] = round(100 * (df_Rooms.RoomType_2020 - df_Rooms.RoomType_2017) 
                             / df_Rooms.RoomType_2017, 2)
df_Rooms['var_lastyear'] = round(100 * (df_Rooms.RoomType_2020 
                                      - df_Rooms.RoomType_2019) 
                                      / df_Rooms.RoomType_2019, 2)
df_Rooms.sort_values(by=['var'])
```

#### QUESTION 2:

The second question: **WHERE** is better to be located?, will compare the number of properties per neighbourhoods in London, how their listings change during the period 2017-2020, and how they change during last year (2020).

For the rest of the questions we will work only with the two predominant room_types, because they represent the
98.4% of the total sample.

```python
# We will work only with the two dominant room_types: Entire home/apt and 
# Private_rooms.
# First step: to filter the listing for only these two types

df_l17 = df_l17[df_l17['room_type'].isin(['Entire home/apt', 'Private room'])]
df_l18 = df_l18[df_l18['room_type'].isin(['Entire home/apt', 'Private room'])]
df_l19 = df_l19[df_l19['room_type'].isin(['Entire home/apt', 'Private room'])]
df_l20 = df_l20[df_l20['room_type'].isin(['Entire home/apt', 'Private room'])]
```

```python
# Working in this question, let's create a dataframe with the number of 
# listings for each neighbourhood for each room_type.
# First we group by every year and then we concatenate all of them.

df_n17 = df_l17.groupby(by=['neighbourhood_cleansed','room_type']) \
               .count()['id'].rename('listings_2017',axis='columns')
df_n18 = df_l18.groupby(by=['neighbourhood_cleansed','room_type']) \
               .count()['id'].rename('listings_2018',axis='columns')
df_n19 = df_l19.groupby(by=['neighbourhood_cleansed','room_type']) \
               .count()['id'].rename('listings_2019',axis='columns')
df_n20 = df_l20.groupby(by=['neighbourhood_cleansed','room_type']) \
               .count()['id'].rename('listings_2020',axis='columns')

df_london20 = pd.concat([df_n17,df_n18,df_n19,df_n20],axis=1)

# Now we create extra columns with some calculations:
#    var = variation in the number of listings during the 2017-2020 period.
#    var_% = same variation in percentage.
#    var_last_year = variation in number of listings during the last year 20.
#    var_%_last_year = same variation in percentage.

df_london20['var'] = df_london20.listings_2020-df_london20.listings_2017
df_london20['var_%'] = round(100 * (df_london20.listings_2020 
                                  - df_london20.listings_2017) 
                                  / df_london20.listings_2017, 2)
df_london20['var_last_year'] = (df_london20.listings_2020 
                              - df_london20.listings_2019)
df_london20['var_%_last_year'] = round(100 * (df_london20.listings_2020
                                            - df_london20.listings_2019) 
                                            / df_london20.listings_2019, 2)
df_london20.sort_values(by=['var_last_year'],ascending=False)
```

```python
# Now we pivot the dataframe to have the neighbourhoods as index and the 
# room_types as columns with the total of listings in the last year 2020.

df_london20_l20 = df_london20.reset_index(level=['neighbourhood_cleansed',
                                                 'room_type'] 
                                         ).pivot(index='neighbourhood_cleansed',
                                                 columns='room_type',
                                                 values='listings_2020')

# The same process but this time with the variation during the period 2017-2020

df_london20_var = df_london20.reset_index(level=['neighbourhood_cleansed',
                                                 'room_type']
                                         ).pivot(index='neighbourhood_cleansed',
                                                 columns='room_type',
                                                 values='var')
df_london20_var['total_varsince2017'] = (df_london20_var['Private room'] 
                                       + df_london20_var['Entire home/apt'])

# Same process again but with the variation for the last year

df_london20_varly = df_london20.reset_index(level=['neighbourhood_cleansed',
                                                   'room_type']
                                        ).pivot(index='neighbourhood_cleansed',
                                                   columns='room_type',
                                                   values='var_last_year')
df_london20_varly['total_varsince2019'] = (df_london20_varly['Private room']
                                         + df_london20_varly['Entire home/apt'])

df_london20_l20
```

```python
# The total of listings of both groups in 2020
total_listing2020 = (np.sum(df_london20_l20['Entire home/apt'])
                   + np.sum(df_london20_l20['Private room']))

# Adding three columns to df_london_l20: the first one calculates the percentage
# of the total listing of every neighbourhood in the global london listing.
# The other two are the series calculated before concatenated to it.

df_london20_l20['%'] =  (100 * (df_london20_l20['Entire home/apt']
                              + df_london20_l20['Private room']) 
                              / total_listing2020)
df_london20_l20 = pd.concat([df_london20_l20, 
                             df_london20_var['total_varsince2017'],
                             df_london20_varly['total_varsince2019']],
                             axis=1)

df_london20_l20 = df_london20_l20.sort_values('%',ascending=False)
df_london20_l20
```

```python
# Finally we just add two new columns with the data of the variations columns 
# but in percentage.

df_london20_l20['var2017%'] = 100 * (df_london20_l20['total_varsince2017'] 
                                     / (df_london20_l20['Entire home/apt']
                                      + df_london20_l20['Private room']
                                      - df_london20_l20['total_varsince2017'])
                                    )
df_london20_l20['var2019%'] = 100 * (df_london20_l20['total_varsince2019']
                                  / (df_london20_l20['Entire home/apt'] 
                                  +  df_london20_l20['Private room'] 
                                  -  df_london20_l20['total_varsince2019'])
                                    )
df_london20_l20
```

**WHAT** is the average price?.
For this question, we'll check the prices per listing grouped by neighbourhoods and the _room_type_.

```python
# This time we will use the price adjusted in the calendar database. We will 
# find the average for each listing in this dataframe
df_c20 = pd.read_csv(path_c20)
```

```python
# Checking the info inside
df_c20.head()
```

```python
# More checking
df_c20.info()
```

```python
# We will work with the adjusted price, so we drop the regular price and 
# convert the dtype to string to after convert it to integer.

df_cal1 = df_c20.drop('price', axis=1)
df_cal1 = df_cal1.astype({'adjusted_price':'string'})
df_cal1.info()
```

```python
# Changing the date column format and data. We only will need for this column 
# the month of each date.
# After that, some renaming and sorting.

df_cal1['date'] = pd.to_datetime(df_cal1['date'])
df_cal1['date'] = df_cal1['date'].dt.month

df_cal1 = df_cal1.rename(columns={'date':'month','adjusted_price':'price',
                                  'listing_id':'id'})
df_cal1 = df_cal1.sort_values(by=['id','month'])
df_cal1

```

```python
# Now drops the 'maximum_nights', not necessary for us at the moment.
# We creates a new dataframe: df_nc20, from the listing database with only the
# data per listing we need, and sorted.

df_cal1 = df_cal1.drop('maximum_nights', axis =1)
df_nc20 = df_l20[['id','neighbourhood_cleansed','room_type','accommodates']] \
                .copy().sort_values(by=['id'])
df_nc20 = df_nc20.rename(columns={'neighbourhood_cleansed':'neighbourhood'})
df_nc20
```

```python
# To convert the 'price' data in floating values, taking only the numerical 
# data from these strings.

df_cal1['price'] = df_cal1['price'].transform(
                        lambda x: float(''.join(re.findall(r'\d+',x)))/100)
df_cal1
```

```python
np.mean(df_cal1[df_cal1['available']=='t'].groupby(by='id').mean())
```

```python
# Merging the data from the listings with the calendar data to receive the 
# final dataframe.

df_calfinal = pd.merge(df_cal1,df_nc20,on=['id'])
df_calfinal
```

```python
# First changing the price to the price per accommodates.
# Second to select only the columns we will need.

df_calfinal['price'] = df_calfinal['price'] / df_calfinal['accommodates']
df_calfinal = df_calfinal[['neighbourhood', 'room_type', 'month', 'available',
                           'price', 'id']]

# Third we take the index from the question before, and apply it to our data 
# to maintain graphically the same order

neighbourhood_index = df_london20_l20.index
sorterIndex = dict(zip(neighbourhood_index, range(len(neighbourhood_index))))
df_calfinal['sort_index'] = df_calfinal['neighbourhood'].map(sorterIndex)
df_calfinal.sort_values('sort_index', ascending=[True], inplace=True)
df_calfinal.drop('sort_index', axis=1, inplace=True)
df_calfinal

```

```python
# Finally we prepare the data to show grouping by neighbourhoods and room_type

df_question3 = df_calfinal.groupby(by=['neighbourhood','room_type'], 
                                   sort=False).mean()
df_question3 = df_question3.reset_index(level=['neighbourhood','room_type'] \
                                       ).pivot(index='neighbourhood',
                                               columns='room_type',
                                               values='price')
df_question3 = df_question3.reindex(df_london20_l20.index)
df_question3
```

```python
# Now one mor step to show the data.

fig3, ax3 = plt.subplots(figsize=(15,5))

# Plotting each kind of room_type
private = ax3.plot(df_question3['Private room'],color = private_color, 
                   label = 'Private Room')
entire = ax3.plot(df_question3['Entire home/apt'],color = entire_color, 
                   label = 'Entire Home/Apt')
ax3.legend()

# Changing some parameters in the line design
ax3.lines[0].set(marker='o', markersize=5, markerfacecolor='white')
ax3.lines[1].set(marker='o',markersize=6)

# Changing title and x-y labels
ax3.set_title('Average price per visitor',**titles_set)
ax3.set_xlabel('Neighbourhoods', **data_labels_set)
ax3.set_ylabel('Average price', **data_labels_set)

# Changing the format in the y axis tick labels
ax3.yaxis.set_major_formatter(StrMethodFormatter('£{x:.2f}'))

# More changes to the ticks labels
ax3.tick_params(axis='y', labelsize=12, right=True, labelright=True)
ax3.tick_params(axis='x', labelsize=13, rotation = 90)

# Erasing the frame
for spine in ax3.spines:
    ax3.spines[spine].set_visible(False)
    ax3.spines[spine].set_visible(False)
    
# Checking the average prices to show them in two different lines
avgprice_entire = df_calfinal[df_calfinal['room_type'] == 'Entire home/apt'
                             ]['price'].mean()
avgprice_private = df_calfinal[df_calfinal['room_type'] == 'Private room'
                              ]['price'].mean()

# Creating the horizontal lines for each average price
ax3.axhline(avgprice_entire, color=entire_color, ls='--')
entire_ann = ax3.annotate('Entire home/apt avg. price: £' 
                          + str(round(avgprice_entire, 2)) 
                          + '/acc.', xy = (8, avgprice_entire), 
                          xytext=(10, avgprice_entire + 30), 
                          arrowprops=dict(arrowstyle='->'))
ax3.axhline(avgprice_private, color = private_color, ls = '--')
private_ann = ax3.annotate('Private room avg. price: £' 
                           + str(round(avgprice_private, 2)) 
                           + '/acc.', xy = (24, avgprice_private), 
                           xytext=(16, avgprice_private + 30), 
                           arrowprops=dict(arrowstyle = '->'))
```

#### QUESTION 4:

```python
# Let's do some calculation:
# First we need the average reviews per listing during the 2017-2018 period:

listing_count_2017 = df_l17.shape[0]
listing_count_2018 = df_l18.shape[0]
avg_listing_jul18 = ((listing_count_2017 + listing_count_2018) 
                    / 2 + dfnan.shape[0])

avg_rev_per_listing_jul18 = (df_rev_jul18_final.n_reviews.sum() 
                            / avg_listing_jul18)
avg_rev_per_listing_jul18
```

```python
# Later we need the average minimum nights per listing. We will use only the 
# 2018 listing for this calc.

avg_min_nights_per_listing_jul18 = (df_listrev17.minimum_nights.mean() 
                                    + df_listrev18.minimum_nights.mean()) / 2
avg_min_nights_per_listing_jul18
#this value on 2014 was 4.6
```

```python
# Now it's time for the total nights calculated in 2018, to compare it with
# the one in the report.

total_nights_jul2018 = airbnb_nights_per_listing2018 * avg_listing_jul18
total_nights_jul2018

# Very close to the 2.2MM in the aribnb 2018 report.
```

```python
# And the bookings per review after that.

bookings_per_rev_jul18 = (nights_per_rev_jul18 / 
                          avg_min_nights_per_listing_jul18)
bookings_per_rev_jul18
```

```python
# The average availability during the period. An aproximation. In total
# offering during a year.

avg_av_365_jul18 = (df_listrev17.availability_365.sum() 
                    + df_listrev18.availability_365.sum())/2
avg_av_365_jul18
```

```python
# For next calculations, we need the occupancy percentage for each kind of
# room_type

av365_privateRooms_jul17 = df_listrev17[df_listrev17['room_type'] == \
                                        'Private room'].availability_365.sum()
av365_entireHome_jul17 = df_listrev17[df_listrev17['room_type'] == \
                                      'Entire home/apt'].availability_365.sum()

av365_privateRooms_jul18 = df_listrev18[df_listrev18['room_type'] == \
                                        'Private room'].availability_365.sum()
av365_entireHome_jul18 = df_listrev18[df_listrev18['room_type'] == \
                                      'Entire home/apt'].availability_365.sum()

avg_av365_privateRooms_jul18 = (av365_privateRooms_jul17 
                                + av365_privateRooms_jul18) / 2
avg_av365_entireHome_jul18 = (av365_entireHome_jul17 
                              + av365_entireHome_jul18) / 2

rev_private = df_rev_jul18_finalb[df_rev_jul18_finalb['room_type'] == \
                                  'Private room'].n_reviews.sum()
rev_entire = df_rev_jul18_finalb[df_rev_jul18_finalb['room_type'] == \
                                 'Entire home/apt'].n_reviews.sum()
total_rev = rev_private + rev_entire

occ_private = (100 * (total_nights_jul2018 * rev_private / total_rev) 
                    / avg_av365_privateRooms_jul18)
occ_entire = (100 * (total_nights_jul2018 * rev_entire / total_rev) 
                    / avg_av365_entireHome_jul18)
```

```python
# Let's start the ML process:

# Then we define our response vector and explanatory matrix:
y = df_selection2['Reviews_per_month']
X = df_selection2.drop(columns=['id','Reviews_per_month'], axis=1)


X.head()
```

```python
X.shape
```

```python
# Let's do ML:

# First we define some constant

test_size = .3
random_value=69

# We have some categorical columns that are no representative and generate
# noise. We will process our linear regression several times filtering some 
# of these noisy columns, checking the number of Trues or '1' they have.
# Defining 'limits' as the list of the sum of those values we will accept in
# each excercise, to drop or not that columns.

limits = [5,10,15,20,25,35,45,55,70,90,120,150,200,500,1000,2500,5000,10000]
    
# Defining the variables we will use to save the results in each excercise

r2_scores_test, r2_scores_train, num_feats, results = [], [], [], dict()

# Now for every cutoff let's implement the Linear Regression

for limit in limits:

    # reduce X matrix, checking the columns that comply with the cutoff limit
    reduce_X = X.iloc[:, np.where((X.sum() > limit) == True)[0]]
    num_feats.append(reduce_X.shape[1])

    # split the data into train and test
    X_train, X_test, y_train, y_test = train_test_split(reduce_X, y, 
                                                        test_size = test_size,
                                                        random_state = \
                                                        random_value)

    # fit the model and obtain predictions response
    lm_model = LinearRegression()
    lm_model.fit(X_train, y_train)
    y_test_preds = lm_model.predict(X_test)
    y_train_preds = lm_model.predict(X_train)

    # append the r2 value from the test set to the lists defined
    r2_scores_test.append(r2_score(y_test, y_test_preds))
    r2_scores_train.append(r2_score(y_train, y_train_preds))
    results[str(limit)] = r2_score(y_test, y_test_preds)

best_limit = max(results, key=results.get)

# Now we repeat the process with the best excercise
# reduce X matrix again
reduce_X = X.iloc[:, np.where((X.sum() > int(best_limit)) == True)[0]]
num_feats.append(reduce_X.shape[1])

# split the data into train and test again
X_train, X_test, y_train, y_test = train_test_split(reduce_X, y, 
                                                    test_size = test_size, 
                                                    random_state=random_value)

# fitting the model again
lm_model = LinearRegression()
lm_model.fit(X_train, y_train)

print(r2_scores_test[np.argmax(r2_scores_test)]) # The best test r-square
print(r2_scores_train[np.argmax(r2_scores_test)]) # Teh best train r-square
print(np.argmax(r2_scores_test)) # the succesful limit implemented
```
